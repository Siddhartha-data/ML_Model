```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

```

```python
dir(iris)
```

```python
iris.target
```

```python
iris.target_names
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head())
```

```python
df['target'] = df['target'].apply(lambda x: iris.target_names[x])
df
```

```python
df[45:55]
```

```python
df[95:105]
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='green', label='setosa')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='blue', label='versicolor')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.legend()
```

```python
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', label='setosa')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='blue', label='versicolor')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.legend()
```

```python
X = df.drop(['target'], axis='columns')
y = df.target
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
X_train
```

```python
X_test
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
model.predict([[6.7, 3.0, 5.2, 2.3]])
```

```python
model_C = SVC(C=1)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_C= SVC(C=10)
model_C.fit(X_train, y_train)
model_C.score(X_test, y_test)
```

```python
model_g = SVC(gamma=15)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
model_g.predict([[5.1, 3.5, 1.4, 0.2]])
```

```python
model_linear_kernel = SVC(kernel='linear')
model_linear_kernel.fit(X_train, y_train)
model_linear_kernel.score(X_test, y_test)
model_linear_kernel.predict([[6.1,	2.9,	4.7,	1.4]])
```

```python
model_linear_kernel = SVC(kernel='poly')
model_linear_kernel.fit(X_train, y_train)
model_linear_kernel.score(X_test, y_test)
model_linear_kernel.predict([[6.0,	3.0,	4.8,	1.8]])
```
